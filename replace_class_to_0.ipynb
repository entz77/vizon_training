```python
from pathlib import Path


def set_first_column_to_zero(folder_path, dry_run=True, encoding="utf-8"):
    """For files not containing '_class_map' in the name, set first token in each line to '0'.

    Lines are expected to be space-separated.
    """
    folder = Path(folder_path)

    if not folder.exists() or not folder.is_dir():
        raise ValueError(f"Invalid folder path: {folder}")

    candidate_files = [
        p for p in sorted(folder.iterdir())
        if p.is_file() and "_class_map" not in p.name
    ]

    print(f"Found {len(candidate_files)} candidate files in {folder}")

    changed_files = []

    for file_path in candidate_files:
        original_lines = file_path.read_text(encoding=encoding).splitlines()
        updated_lines = []
        file_changed = False

        for line in original_lines:
            stripped = line.strip()
            if not stripped:
                updated_lines.append(line)
                continue

            parts = stripped.split()
            if not parts:
                updated_lines.append(line)
                continue

            if parts[0] != "0":
                parts[0] = "0"
                file_changed = True

            updated_lines.append(" ".join(parts))

        if not file_changed:
            print(f"No change: {file_path.name}")
            continue

        changed_files.append(file_path.name)

        if dry_run:
            print(f"[DRY RUN] Would update: {file_path.name}")
        else:
            file_path.write_text("\n".join(updated_lines) + "\n", encoding=encoding)
            print(f"Updated: {file_path.name}")

    print(f"Total files changed: {len(changed_files)}")
    return changed_files


# Example usage:
labels_folder = "datasets/softener_pewangi/labels"
# changed_preview = set_first_column_to_zero(labels_folder, dry_run=True)

# Apply changes:
set_first_column_to_zero(labels_folder, dry_run=False)
```
